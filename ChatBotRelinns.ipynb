```python
# Checking Working Condition
data = bs4.BeautifulSoup(link, 'lxml')
data
```

```python
# Finding all paragraphs ('<p>') within the 'data'
dataParagraph = data.find_all('p')

# Initializing an empty string to store concatenated text from paragraphs

dataText = ''

# extracting text, and concatenating it

for i in dataParagraph:
    dataText += i.text
# Converting the concatenated text to lowercase

dataText = dataText.lower()

len(dataText)
```

```python
# Replacing square brackets with numbers inside 'dataText' with a space if Any
dataText = re.sub(r'\[[0-9]*\]', ' ', dataText)
```

```python
# nltk.download("all")
```

```python
# Tokenizing the text into sentences using nltk
sen = nltk.sent_tokenize(dataText)

# Tokenizing the text into words using nltk
words = nltk.word_tokenize(dataText)

sen
```

```python
temp = []  # Creating an empty list 'temp' to store sub-sentences temporarily

# Iterating through each sentence in 'sen'
for sentence in sen:
    # Splitting each sentence into sub-sentences based on periods ('.')
    for subSen in sentence.split('.'):
        # Checking if the sub-sentence is not an empty string
        if subSen != '':
            # Appending non-empty sub-sentences to the temporary list 'temp'
            temp.append(subSen)

# Assigning the content of 'temp' back to 'sen', effectively updating 'sen' with the sub-sentences
sen = temp[:]
```

```python
from nltk.stem import WordNetLemmatizer


def performLemma(tokens):
    # Lemmatizing each token in the given list of tokens
    return [WordNetLemmatizer().lemmatize(token) for token in tokens]
```

```python
# Creating a dictionary to replace punctuation marks with None
pr = dict((ord(punctuation), None) for punctuation in string.punctuation)
```

```python
def getProcessedText(document):
    # Converts the document to lowercase, removes punctuation using the 'pr' dictionary,
    # tokenizes the text into words using nltk, and then lemmatizes the tokens
    return performLemma(nltk.word_tokenize(document.lower().translate(pr)))
```

```python
greetingInputs = ('Hey', 'Good Morning', 'Good Evening', 'Hi')
greetingResponse = ['hey', 'How are you?', 'Hello, How you doing']


def generateGreetingResponse(greeting):
    for token in greeting.split():
        if token.lower() in greetingInputs:
            return random.choice(greetingResponse)
    # If none of the tokens match, return None or an appropriate response
```

```python
# Importing TF-IDF Vectorizer from Scikit-learn
from sklearn.feature_extraction.text import TfidfVectorizer
# Importing cosine similarity from Scikit-learn
from sklearn.metrics.pairwise import cosine_similarity


def generateResponse(userInput):
    botResponse = ''  # Initializing the bot's response variable
    # Appending the user's input to the list of sentences (sen)
    sen.append(userInput)

    # Creating a TF-IDF Vectorizer with customized tokenizer and stop words
    wordVec = TfidfVectorizer(tokenizer=getProcessedText, stop_words="english")

    # Fitting the TF-IDF Vectorizer to the sentences and transforming them into vectors
    wordVectors = wordVec.fit_transform(sen)

    # Calculating cosine similarity between the last sentence and all other sentences
    similarVectorValues = cosine_similarity(wordVectors[-1], wordVectors)

    # Finding the index of the most similar sentence to the user's input
    similarSentenceNumber = similarVectorValues.argsort()[0][-2]

    # Flattening and sorting the similarity values to find the most similar one
    matchedVector = similarVectorValues.flatten()
    matchedVector.sort()
    vectorMatched = matchedVector[-2]

    # Checking if there's a meaningful match or defaulting to an "I Didn't Understand" response
    if vectorMatched == 0:
        botResponse = botResponse + "Sorry, I Didn't Understand"
        return botResponse
    else:
        # Generating bot's response based on the most similar sentence
        botResponse = botResponse + sen[similarSentenceNumber]
        return botResponse
```

```python
flag = True  # Flag to control the conversation loop
print("Hello! I am ChatBot. Please ask your Question\nIf You Don't Have Questions Press Q.")

while flag:
    theInput = input()  # Getting user input
    theInput = theInput.lower()  # Converting input to lowercase for uniformity

    if theInput == 'q':
        print("Nice Talking To You.")
        flag = False  # Setting flag to False to exit the conversation loop

    elif generateGreetingResponse(theInput) != None:
        # Checking if the input contains a greeting; if so, generating a greeting response
        print(f"Bot: {generateGreetingResponse(theInput)}")

    else:
        # If the input is not a greeting, generating a response based on the input
        print(f"Bot: {generateResponse(theInput)}")
        # Removing the input from the list of sentences (sen)
        sen.remove(theInput)
```
